```python
import os
import pandas as pd
import numpy as np
```

```python
train_path = os.path.join('..','data','raw','train.csv')
train_df = pd.read_csv(train_path)
train_df
```

```python
train_df_copy = train_df.copy()
train_df_copy
```

```python
train_df_copy['keyword'] = ['unknown' if type(i) == float else i for i in train_df_copy.keyword.values]
train_df_copy['location'] = ['unknown' if type(i) == float else i for i in train_df_copy.location.values]
```

```python
train_df_copy
```

```python
train_df_copy['corpus'] = train_df_copy.keyword + " " + train_df_copy.location + " " + train_df_copy.text
train_df_copy = train_df_copy[['id','location','keyword','text','corpus','target']]
train_df_copy
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = train_df_copy.corpus.to_list()
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
# print(vectorizer.get_feature_names())
# print(X)
```
